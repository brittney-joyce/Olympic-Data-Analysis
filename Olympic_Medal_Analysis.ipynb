```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import seaborn
```

```python
#read csv
medals = "../TM Group Project/olympic_medalists.csv"
medals_df = pd.read_csv("olympicsall.csv", encoding = "ISO-8859-1")
```

```python
medals_df.head()
```

```python
#countries with the most medals
country_count = medals_df[["Country", "Medal"]]
country_count.head(100)
```

```python
country_medals= pd.DataFrame(country_count[("Country")].value_counts())
country_medals.head()
```

```python
# Bar graph for Country Medal Count
x_axis = np.arange(len(country_medals))
tick_locations = [value+0.4 for value in x_axis]
#plt.figure(figsize=(500,1))
plt.bar(x_axis, country_medals["Country"], color='r', alpha=0.5, align="edge")
plt.xticks(tick_locations, country_medals["Country"], rotation="vertical")

plt.xlim(-0.25, 100)
plt.ylim(0, 2000)

plt.show()
```

```python
#gender analysis by year 
gender_analysis = pd.DataFrame( {"Medals Count":medals_df.groupby(['Gender','Year', 'Season']).size()} ).reset_index() #.groupby(['Gender','Year']) 
men = gender_analysis[gender_analysis["Gender"] == "Men"] 
women = gender_analysis[gender_analysis["Gender"] == "Women"]

gender_analysis.to_csv('gender_analysis.csv')
```

```python
#gender analysis by year
gender_analysis = pd.DataFrame( {"Medals Count":medals_df.groupby(['Gender','Year', 'Season']).size()} ).reset_index() #.groupby(['Gender','Year'])
men = gender_analysis[gender_analysis["Gender"] == "Men"]
women = gender_analysis[gender_analysis["Gender"] == "Women"]

gender_analysis.head()
```

```python
## Men vs Women Scatter Plot
plt.scatter(men['Year'], men['Medals Count'], c="blue", edgecolor="black", linewidths=1, marker="o", label="Men")
plt.scatter(women['Year'], women['Medals Count'], c="pink", edgecolor="black", linewidths=1, marker="o", label = "Women")

plt.title("Medal Count By Gender")
plt.ylabel("Medal Count")
plt.xlabel("Olympic Year")
plt.xlim((1890,2020))
plt.grid(True)

plt.legend(fontsize="small", mode="Expanded", 
                  numpoints=1, scatterpoints=1, 
                  loc="upper left", title="Gender", 
                  labelspacing=0.5)

plt.show()
```

```python
## Men vs Women Line Graph
menplot= plt.plot(men['Year'], men['Medals Count'], c="blue",label="Men")
womenplot= plt.plot(women['Year'], women['Medals Count'], c="pink", label = "Women")

plt.title("Medal Count By Gender")
plt.ylabel("Medal Count")
plt.xlabel("Olympic Year")
plt.xlim((1890,2020))
plt.grid(True)

plt.legend(fontsize="small", mode="Expanded", 
                  numpoints=1, scatterpoints=1, 
                  loc="upper left", title="Gender", 
                  labelspacing=0.5)

plt.show()
```
