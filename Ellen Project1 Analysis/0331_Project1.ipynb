```python
# Calculate the number of unique country in the DataFrame
country_count = len(gpa_df["Country"].unique())

# Calculate the earliest/latest year a medal was won
earliest_year = gpa_df["Year"].min()
latest_year = gpa_df["Year"].max()

# Calculate the total medals for the entire dataset
total_medal = gpa_df["Medal"].count()
```

```python
# Place all of the data found into a summary DataFrame
summary_table = pd.DataFrame({"Total Unique Country":country_count,
                             "Earliest Year":[earliest_year],
                             "Latest Year":[latest_year],
                             "Total Medals":[total_medal]})
summary_table
```
